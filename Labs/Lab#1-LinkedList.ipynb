```c++
template<class T>
class LinkedList
{
    private:
        Node<T> * next;
    public:
        LinkedList();
        
        void pushFront(const T &value);
        T topFornt()const;
        bool popFront()const;
    
        void pushBack(const T &value);
        T topBack()const;
        bool popBack()const;
    
        void insert(const T &value);
        T find(int index);//[]
    
        long size()const;
        bool isEmpty()const;
    
        
        ~LinkedList();
}
template <class T>
void LinkedList::pushFront(const T & value)
{
    Node<T> * p;
    p -> key = value;
    if (head == NULL)
        head = p;
    else
    {
        p -> next = head;
        head = p;
    }
}

template <class T>
bool LinkedList::popFront()const
{
    if (head == NULL)
        return false;
    else
    {
        Node * temp;
        
        temp = head;
        head = head -> next;
        
        delete (temp);
        return true;
    }
}

template <class T>
T LinkedList::topFront()const
{
    return head -> key;
}

template <class T>
void LinkedList::pushBack(const T & value)
{
    Node<T> *p;
    p->key = value;
    
    if(head == NULL)
        head = p;
    else
    {
        Node<t> *temp=head;
        while(temp -> next != NULL)
            temp = temp -> next;
        temp -> next = p;
    }
}

template <class T>
bool LinkedList::popBack()const
{
    if (head == NULL)
        return false;
    else
    {
        Node * temp;
        
        temp = head;
        head = head -> next;
        
        delete (temp);
        return true;
    }
}

template <class T>
T LinkedList::topBack()const
{
    return head -> key;
}

template <class T>
long LinkedList::size()const
{
    long counter = 0;
    
    Node<T> *temp = head;
    while (temp != NULL)
    {
        temp = temp -> next;
        counter ++;
    }
    return counter;
}

template <class T>
bool LinkedList::isEmpty()const
{
    return head == NULL;
}
```
